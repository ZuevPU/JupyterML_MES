```python
import csv
import matplotlib.pyplot as plt

f = open('c:/data2.csv')

data_1 = []
data_2 = []
data_3 = []
data_4 = []

data = csv.reader(f)

for row in data:
    data_1.append(int(row[0]))
    data_2.append(int(row[1]))
    data_3.append(int(row[2]))
    data_4.append(int(row[3]))

x=np.arange(len(data_1))+1

plt.bar(x,data_1,0.2,color='red')
plt.bar(x+0.25,data_2,0.2,color='green')
plt.bar(x+0.5,data_3,0.2,color='blue')
plt.bar(x+0.75,data_4,0.2,color='orange')

plt.xticks(x + 0.5*0.5, x)
    
f.close()
```
